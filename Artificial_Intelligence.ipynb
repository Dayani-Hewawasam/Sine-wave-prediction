```python
import numpy as np  # For sine wave generation & numerical operations
import pandas as pd  # (Optional) If working with structured data
import matplotlib.pyplot as plt  # For plotting sine waves
import math
import torch
import torch.nn as nn
```

```python
x = torch.linspace(0,799,800)
```

```python
x[-10:]
```

```python
y = torch.sin(x * 2*math.pi/40)
```

```python
plt.figure(figsize=(12, 4))
plt.xlim(-10,805)
plt.grid(True)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Sine plot")
plt.plot(y.numpy(),color='b')
plt.show()
```
